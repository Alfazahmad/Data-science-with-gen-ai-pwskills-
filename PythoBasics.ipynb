Theory Questions:

1. What is Python, and why is it popular?
Python is a high-level, interpreted programming language known for its simplicity and readability. It is popular due to its clean syntax, large library ecosystem, ease of learning, and wide usage in fields such as web development, data science, artificial intelligence, and automation.

2. What is an interpreter in Python?
An interpreter in Python translates the Python code line by line into machine code, making it executable. It runs the program directly without compiling it into machine code beforehand, which helps with quick testing and debugging.

3. What are pre-defined keywords in Python?
Pre-defined keywords in Python are reserved words that have special meaning in the language syntax, such as if, else, for, while, def, etc. They cannot be used as variable names.

4. Can keywords be used as variable names?
No, keywords cannot be used as variable names in Python, as they are reserved for defining the structure of the language.

5. What is mutability in Python?
Mutability refers to the ability of an object to be changed after its creation. Mutable objects can have their contents modified, while immutable objects cannot be changed once they are created.

6. Why are lists mutable, but tuples are immutable?
Lists are mutable because they are designed to allow dynamic changes, such as adding, removing, or altering elements. Tuples, on the other hand, are immutable to prevent modification, making them suitable for fixed data that should not be changed.

7. What is the difference between "==" and "is" operators in Python?

== checks if the values of two objects are equal.

is checks if two objects refer to the same memory location (i.e., they are the same object in memory).


8. What are logical operators in Python?
Logical operators in Python include and, or, and not, used to combine conditional statements and evaluate boolean expressions.

9. What is type casting in Python?
Type casting is the process of converting one data type into another, such as converting an integer to a float or a string to an integer.

10. What is the difference between implicit and explicit type casting?

Implicit type casting (or type coercion) is done automatically by Python when it converts a smaller data type to a larger one (e.g., int to float).

Explicit type casting requires the user to manually convert one type to another using functions like int(), float(), etc.


11. What is the purpose of conditional statements in Python?
Conditional statements (if, elif, else) are used to execute certain blocks of code based on specific conditions, enabling decision-making in a program.

12. How does the elif statement work?
The elif (else if) statement allows checking multiple conditions. If the if condition is false, Python checks the elif conditions in order until one evaluates to True.

13. What is the difference between for and while loops?

A for loop is used for iterating over a sequence (e.g., list, range).

A while loop repeatedly executes a block of code as long as a given condition is true.


14. Describe a scenario where a while loop is more suitable than a for loop.
A while loop is more suitable when the number of iterations is not predetermined, and you want the loop to run until a certain condition is met (e.g., waiting for user input).
